#ANALYSING DATA BY GRAPHS

```python
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
x = data['Region']
```

```python
y=data['Estimated Unemployment Rate (%)']
```

#ANALYSING DATA WITH BAR GRAPH

```python
fg = px.bar(data,x='Region',y='Estimated Unemployment Rate (%)',color='Region',
            title='Unemploymeny Rate (State Wise) by Bar Graph',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

#ANALYSING DATA WITH BOX PLOT

```python
fg = px.box(data,x='Region',y='Estimated Unemployment Rate (%)',color='Region',
            title='Unemploymeny Rate (Statewise) by Box Plot',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

#ANALYSING DATA WITH SCATTER PLOT

```python
fg = px.scatter(data,x='Region',y='Estimated Unemployment Rate (%)',color='Region',
                title='Unemploymeny Rate (Statewise) by Scatter Plot',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

#ANALYSING DATA WITH HISTOGRAM

```python
fg = px.histogram(data,x='Region',y='Estimated Unemployment Rate (%)',color='Region',
                  title='Unemploymeny Rate (Statewise) by Histogram',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```
